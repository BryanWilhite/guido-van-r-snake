In Python, Booleans are title-cased as:

```python
True
False
```

Python is case-sensitive. It will throw an error for a Boolean expression like:

```python
true
```

Python has [Boolean operations](https://docs.python.org/2/library/stdtypes.html#boolean-operations-and-or-not) and [comparison operations](https://docs.python.org/2/library/stdtypes.html#comparisons) (which return Booleans).

Comparison operations:

```python
a = 3
b = 5

a == b
a != b
a > b
a < b
```

Comparison operations can indirectly convert numbers to Booleans. This can be done explicitly by constructing with the Boolean object, `bool`:

```python
bool(28)
bool(-2.711828) # returns True
bool(0)
```

Python returns `True` for all numbers except _0_.

> In Python, non-trivial values are converted to `True` while trivial ones are converted to `False`.

The `bool` constructor takes `object` as its argument, boxing anything, including strings:

```python
bool("Turing")
bool(" ") # returns True
bool("")
```

Booleans can be converted to other types: 

```python
str(True)
int(False)
int(True)
5 + True
10 * False
```
