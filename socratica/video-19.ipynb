Python offers a pseudo-random number generator in the `random` module [ðŸ“– [docs](https://docs.python.org/3/library/random.html)]:

```python
import random
```

```python
dir(random)
```

```python
for i in range(10):
    print(random.random()) # interval [0, 1)
```

The documentation warns us:

>The pseudo-random generators of this module should not be used for security purposes. For security or cryptographic uses, see the `secrets` module.

We see that `.random()` pulls from the interval $[0, 1)$ by default. Let us specify a different interval with our own function:

```python
def my_random(floor, max):
    interval = max - floor
    return interval * random.random() + floor
```

```python
for i in range(10):
    print(my_random(3, 7)) # interval [3, 7)
```

Python gives us the `.uniform()` function as the equivalent of our function above:

```python
help(random.uniform)
```

```python
for i in range(10):
    print(random.uniform(3, 7)) # interval [3, 7)
```

The reason why `.uniform()` has its name is to remind us that number genration is within a [uniform distribution](https://en.wikipedia.org/wiki/Uniform_distribution_(continuous)) of possible values. The distribution of `.random()` is also uniform.

For a [normal distribution](https://en.wikipedia.org/wiki/Normal_distribution), we have the `.normalvariate()` function:

```python
help(random.normalvariate)
```

### Discrete Probability Distributions
    
To generate `int` instead of `float` numbers, we have [discrete distributions](https://en.wikipedia.org/wiki/List_of_probability_distributions#Discrete_distributions) generated by `.randint()`:

```python
help(random.randint)
```

```python
for i in range(20):
    print(random.randint(1, 6))
```

To randomly select an item from a sequence, we have `.choice()`:

```python
help(random.choice)
```

```python
for i in range(10):
    print(random.choice(['rock', 'paper', 'scissors']))
```

```python
for i in range(10):
    print(random.choice(('rock', 'paper', 'scissors')))
```

```python
for i in range(10):
    print(random.choice({0:'rock', 1:'paper', 2:'scissors'}))
```

```python
for i in range(10):
    print(random.choice(set(['rock', 'paper', 'scissors'])))
```

_Reminder:_ a `set` is not a sequence.
