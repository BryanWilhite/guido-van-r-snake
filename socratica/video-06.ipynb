In Python culture, when a numeric type has more length (range) than another numberic type, we say that type is _wider_ than the other. Conversely, when a numeric type has less range, it is more _narrow_. For example, the video above shows that, in Python 2.x, longs are more narrow than floats:

```python
>>> float1 = 3.14
>>> long(float)
3L
```

In Python 2.x, floats are wider than longs. But in both Python 2.x and 3.x, complex numbers are wider than floats:

````python
float1 = 3.14
complex1 = float1 + 0j     # converting to `complex` by addition
complex2 = complex(float1) # converting to `complex` by class construction

complex1 == complex2 # should return `True`

float(complex1) # should throw an error
````

We can say that this error was thrown because there is no convention in place to _narrow_ down a complex number into an integer in Python. On the other hand, Python has a convention for arithmatic operations with different numeric types:

>widen the narrow type and then perform the arithmatic operation

In Python 2.x, it follows that:

```plaintext
<int> + <long> = <long>
```

For the other numeric types:

````python
a = 2
c = 6.0
d = 12 + 0j
````

````python
a + c
````
